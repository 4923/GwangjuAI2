```python
# Q1 변수의 선언, 변수 자료형의 소개
def variable_confirm( values_list ) : # 변수 자료형을 분석해주는 함수 선언
  for i in range( len(values_list) ) : # 변수별로 분석하기 위한 반복문
      print( f'변수 값: {values[i]}' ) # 변수의 값
      print( '변수 자료형 : {} \n'.format( type(values[i])) ) # 변수의 자료형

values = [3, 11.3, "Gwangju", {"음식가격" : 3000}, [1,2,3], (33, 44, 55)] # 분석을 위해 선언한 리스트
variable_confirm(values) # 리스트를 실행한 함수
```

```python
# Q2 리스트와 리스트 기능 활용
def list_confirm( values_list ) : # 리스트를 분석해주는 함수 선언
  
  name_masking = "" # 이름 마킹 처리를 하게 될 변수 선언
  for i in range( len(values_list[0])-1 ) : # 이름 마스크 처리를 위한 반복문
    name_masking += "*" # 이름 마킹에 성을 제외한 이름을 * 처리 하기 위해 텍스트 삽입

  # 0번째 데이터에서 이름을 마킹 처리
  values_list[0] = values_list[0].replace( values_list[0][1:len(values_list[0])], name_masking )
  # 1번째 데이터에서 휴대폰 번호를 마킹 처리
  values_list[-1] = values_list[-1].replace( values_list[1][-4:len(values_list[1])], "****" )
  
  return values_list # 이름, 휴대폰 번호 마킹 처리된 리스트 배출

values = ["박선호", "010-1234-5678"] # 분석을 위해 선언한 변수들
list_confirm(values) # 변수를 실행한 함수
```
