```python
# Q1 변수의 선언, 변수 자료형의 소개

def check_type(values):     # list
    for value in values:
        print(f"변수 값: {value}")
        print(f"변수 자료형 : {type(value)}\n")

values = [3, 11.3, "Gwangju", {"음식가격" : 3000}, [1,2,3], (33, 44, 55)]
check_type(values)
```

```python
# Q2 리스트와 리스트 기능 활용
# ToDo : 이름과 번호 뒷자리 마킹처리

def list_confirm(values):
    name = values[0][0] + '**'
    phone_number = values[1][:-4] + "****"
    return [name, phone_number]

values = ["박선호", "010-1234-5678"] # 분석을 위해 선언한 변수들
list_confirm(values) # 변수를 실행한 함수
```

```python
import time
import os
# Q3 반복문과 조건문 활용

Life = 1 # Life 변수 선언
questions = ["진짜 추워의 반대말은?", "송해 할아버지가 샤워를 하면?", "가장 억울한 도형은?"]
answers = ["아마추어", "뽀송뽀송해", "원통"]

for idx, question in enumerate(questions):
    # user input
    print(f"Q{idx+1}. {question} : ", end="")
    answer = input()
    print(answer)

    # answer check
    status = False
    if answer.strip(" ") == answers[idx]: status = True

    # return
    print(f"A{idx+1}. ", end="")
    if status: print("정답입니다.")
    else: print("틀렸습니다.")
```
