# Loss Function

## loss function : regression

### MAE

> Mean Squared Error

$$ \frac{1}{N} \sum_{i=1}^{N} (y_{i} - \hat{y_{i}})^{2}$$

실제 값 $y_{i}$ 과 예측 값 $\hat{y_{i}}$의 차이를 제곱한 값들의 평균 \frac{1}{N} \sum_{i=1}^{N}이다.  
$\therefore$ **평균 제곱 오차** 라고도 한다.

- **Squared**
    1. 음수를 무시하기 위함.
    2. 작은 오차를 확대하여 관찰하기 위함
        - 오차가 실제보다 커져 왜곡이 생길 수 있으며 이를 보정하기 위해 고안된 것이 $RMSE$다.

```python
x = np.arange(1, 6, 1)
y = np.array([10, 20, 30, 40, 50])
y_hat = np.array([5, 24, 23, 37, 59])

output_num = len(y)
loss = y - y_hat 
mse = sum(loss**2) / output_num

print("MSE : ", mse)
```

```python
# visualization 

plt.subplot(121)
plt.plot(x, y, 'bo', label="label")
plt.plot(x, y_hat, 'g-o', label="predict")
plt.grid()
plt.legend()
plt.xlabel('input')
plt.ylabel('output')

plt.subplot(122)
plt.plot(x, loss, 'm--o', label="loss")     # origin loss
plt.plot(x, loss**2, 'r--o', label = "squared loss")      # mse
plt.xlabel('input')
plt.ylabel('loss')
plt.legend()
plt.grid()
plt.show()
```

제곱한 값이 더 크다는 것을 확인할 수 있다. 

예측값과 실제값의 차이의 면적을 합한 값이며 이상치에 영향을 받는다. ($\because$ 제곱값이므로)

$RMSE$
> Root MSE

$$ \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_{i} - \hat{y_{i}})^{2}} $$

손실이 급격하게 커지는 상황에 적합하다.

$MAE$

> Mean Absolute Error

$$ \frac{1}{N} \sum_{i=1}^{N} \lvert y_{i} - \hat{y_{i}}\rvert $$

MSE가 에러 제곱의 평균을 구했다면 MAE는 **에러 절대값의 평균**을 구한다.

- MSE에서 음수를 처리하기 위해 제곱을 취했다면 MAE는 절댓값을 취한다.
- 이상치가 많을 때 적합하다.
