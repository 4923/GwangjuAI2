```python
import csv
```

```python
# 데이터를 여는 것과 읽는 것은 다른 개념

with open('dataset/dataset/abalone_mini.csv') as csvfile:   # 열기
    csvreader = csv.reader(csvfile)     # 읽기

    # 반복문으로 데이터를 실제 변수에 할당
    rows = []
    for row in csvreader:
        rows.append(row)
```

```python
rows
```

```python
# hint : zeros, enumerate

abalone_mini_buffer = np.zeros((len(rows), 2+len(rows[0])))
# abalone_mini_buffer.shape   (5, 11)

for idx, row in enumerate(rows):
    # one hot vector
    if row[0] == 'M': abalone_mini_buffer[idx][0] = 1
    elif row[0] == 'F': abalone_mini_buffer[idx][1] = 1
    elif row[0] == 'I': abalone_mini_buffer[idx][2] = 1

    # 남은 데이터 병합
    abalone_mini_buffer[idx][3:] = row[1:]

abalone_mini_buffer
```

### 신경망을 동작할 때 가장 중요한 것?

weight, bias
- weight : 입력값의 개수만큼 필요  
- bias : 연산횟수만큼 필요

### 함수 안에 함수 작성하기?

위 방법처럼 메인 함수 안에 다른 함수를 실행시켜 모든 프로세스를 한번에 동작시킨다.

```python
RND_MEAN, RND_STD = 0, 1
input_cnt, output_cnt = 10, 1


def import_data():
    pass 

def model_init():
    global weight, bias     # 전역변수 선언
    # 가중치 / 편향
    # 파라미터 -> 무작위 값으로 설정
    weight = np.random.normal(RND_MEAN, RND_STD, size = [input_cnt, output_cnt])     # np.random.normal : 평균, 표준편차, 개수
    bias = np.random.normal(RND_MEAN, RND_STD, size = output_cnt)   # 개수가 output_cnt임에 유의

# 순전파 구현 실습
# 구해야 할 것 : y_hat = (x * weight) + bias 
def train_and_test(x):  # x는 one hot encoding한 `data`의 처음부터 종속변수 전까지만 [:-1] 대입한다.
    y_hat = np.matmul(x, weight) + bias
    return y_hat



def main_exec(x):    # main function
    import_data()   # func1
    model_init()    # func2
    train_and_test(x)    # func3

# exec : y_pred
y_pred = main_exec(abalone_mini_buffer[:, :-1])
print(y_pred)

# exec : weight, bias
print(f'weight\n{weight}')
print(f'bias : {bias}')
```
