```python
# LV1 : 데이터셋에 빈 값이 존재합니다. 확률적인 분포(백분율) 을 기반으로 데이터셋을 새롭게 생성하세요!

def menu_pick(RANK, pick):
  menu = np.empty([RANK], dtype=object)
  prob = list(np.random.randn(pick))
  for idx in range(RANK):
    menu[idx] = input(f"{idx+1}번째로 좋아하는 음식은? : ")
  your_pick = np.random.choice(menu, pick, prob)
  return your_pick

print(menu_pick(3, 10))
```

```python
# LV2 : 로또 번호 생성기

def lotto():
  lotto = np.unique(np.random.randint(1, 46, 7))
  while lotto.size != 7:
    new = np.random.randint(1, 46, 1)
    print(f"\t중복된 값이 있으므로 숫자를 새로 뽑습니다. -> new number : {new}")
    lotto = np.unique(np.concatenate([lotto, new]))
  return lotto

print(f"생성된 로또 번호는 {lotto()} 입니다")
```
